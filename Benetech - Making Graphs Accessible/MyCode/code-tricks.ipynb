# 使用 pandas json_normalize 进行json解耦

```python


def load_annotation(name):
    with open(name) as f:
        annotation = json.load(f)
        annotation = json_normalize(annotation)
    return annotation

def get_annotation_df(annotations_path):
    annotation_df = pd.DataFrame()
    for name in os.listdir(annotations_path)[:10]:
        annotation_df = pd.concat([annotation_df, 
                                   load_annotation(Path(annotations_path) / name)], axis = 0)
    annotation_df.reset_index(drop=True, inplace=True) # 因为是concat出来的，index都为0，如果不reset，后面merge就会失败
    return annotation_df
            
def get_chart_type_counts(annotations):
    chart_type_list = [chart.get('chart-type') for chart in annotations]
    return np.unique(chart_type_list, return_counts=True)


def flatten_annot_df(annot_df, JSON_COLUMNS):    
    for column in JSON_COLUMNS:
        column_as_df = json_normalize(annot_df[column])
        column_as_df.columns = [f"{column}.{subcolumn}" for subcolumn in column_as_df.columns]
        annot_df = annot_df.drop(column, axis=1).merge(column_as_df, right_index=True, left_index=True)
    return annot_df

 
x = get_annotation_df(root_dir / "train/annotations")

JSON_COLUMNS = [column for column in x.columns 
                if type(x.loc[:, column].values[0]) == list]
#flatten_annot_df(x, JSON_COLUMNS)
flatten_annot_df(x, JSON_COLUMNS).to_csv(root_dir / "test.csv", index = False)
# get_annotation_df(root_dir / "train/annotations")
```

```python
def extract_plot_bb(data):
    x0 = data["plot-bb"]["x0"]
    y0 = data["plot-bb"]["y0"]
    w = data["plot-bb"]["width"]
    h = data["plot-bb"]["height"]
    return [x0, y0, w, h, "plot-bb"]


def extract_label_bbox(data, rows, img_id):
    common = []
    common+= [img_id, data["source"], data["chart-type"]]
    for box in data["text"]:
        x0 = box["polygon"]["x0"]
        y0 = box["polygon"]["y0"]
        w = box["polygon"]["x1"] - box["polygon"]["x0"]
        h = box["polygon"]["y3"] - box["polygon"]["y0"]
        label = box["role"]
        box_row = common + [x0, y0, w, h, label]
        rows.append(box_row)
    rows.append(common + extract_plot_bb(data)) 
    return rows
        
def extract_tick_bbox(data, rows, img_id):
    common = []
    common+= [img_id , data["source"], data["chart-type"]]
    for axis in data["axes"].keys():
        for box in data["axes"][axis]["ticks"]:
            x0 = box["tick_pt"]["x"] - 5
            y0 = box["tick_pt"]["y"] - 5
            w = 10
            h = 10
            label = axis + "-tick"
            box_row = common + [x0, y0, w, h, label]
            rows.append(box_row)
    return rows
```
